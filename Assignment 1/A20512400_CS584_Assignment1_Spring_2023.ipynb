1.  Explain the differences between supervised, semi-surpervised, weakly-supervised and unsupervised learning.




**Answer:** 

Supervised learning: In this type of machine learning, the model is trained with labeled data, which means both input and output data is provided to the model during the training phase. The model learns to map inputs to outputs by finding the relationship between the two. This approach is used for classification and regression problems.

Semi-supervised learning: In this type of learning, the model is trained with both labeled and unlabeled data. The goal is to make the best use of limited labeled data and large amounts of unlabeled data to improve the model's accuracy. This approach is used when acquiring labeled data is expensive or time-consuming.

Weakly-supervised learning: In this type of learning, the model is trained with weakly labeled data, which means the data is only partially labeled. For example, image classification problems where the label may only indicate the presence of an object in the image and not its exact location. This approach is used when acquiring precise labels is difficult or infeasible.

Unsupervised learning: In this type of learning, the model is trained with only input data, without any corresponding output labels. The goal is to find patterns or relationships in the data without being told what the output should be. This approach is used for clustering, dimensionality reduction, and anomaly detection problems


2. Explain the difference between regression and classification.




**Answer:**
Regression and classification are two common types of supervised machine learning tasks.

Regression is a type of prediction problem where the goal is to predict a continuous output value. For example, predicting the price of a house based on its size, location, and other features. The output in regression is a continuous real value, such as a floating point number.

Classification, on the other hand, is a type of prediction problem where the goal is to predict a categorical output value. For example, predicting whether an email is spam or not, or predicting the species of an iris plant based on its physical characteristics. The output in classification is a categorical value, represented as a class label.

So, the main difference between regression and classification is that in regression, the output is a continuous value, while in classification, the output is a categorical value.



3. Explain how to perform N-fold cross-validation.




**Answer:**
N-fold cross-validation is a technique used to evaluate the performance of machine learning models. It is a model validation technique that helps to ensure that the model is not overfitting or underfitting the data. The idea behind N-fold cross-validation is to divide the training data into N partitions or folds, and then train the model N times, using N-1 folds as the training data and the remaining fold as the validation data. After each iteration, the model is tested against the validation data, and the performance is recorded. Finally, the average performance across all N iterations is used to evaluate the model.

The steps to perform N-fold cross-validation are as follows:

1 -> Divide the training data into N folds.

2 -> For each iteration, choose one of the N folds as the validation data and use the remaining N-1 folds as the training data.

3 -> Train the model on the training data.

4 -> Evaluate the model on the validation data.

5 -> Repeat steps 2 to 4 N times, each time using a different fold as the validation data.

6 -> Average the performance of the model across all N iterations to obtain the final performance estimate.

N-fold cross-validation is a useful technique for improving the accuracy and stability of machine learning models. It allows you to get a good estimate of model performance on unseen data and helps you to avoid overfitting or underfitting the data.

John owns a retail store for selling phones. The phones are manufactured at three different factories, A, B,

number of phones being sold at John’s store is 10,000. One day, a customer walks up to John’s store, and ask
for a refund for a defective phone.
1. What is the probability of a phone being defective?
2. What is the probability that this defective phone is manufactured at factory A?
3. What is the probability that this defective phone is manufactured at factory B?
4. What is the probability that this defective phone is manufactured at factory C?




**Answer:**
Based on the information given, we have:

P(A) = 20% = 0.2

P(B) = 30% = 0.3

P(C) = 50% = 0.5

P(D | A) = 2% = 0.02 (probability that a phone is defective given that it was manufactured at factory A)

P(D | B) = 1% = 0.01 (probability that a phone is defective given that it was manufactured at factory B)

P(D | C) = 0.6% = 0.006 (probability that a phone is defective given that it was manufactured at factory C)


1. The probability of a phone being defective: P(D) = P(D and A) + P(D and B) + P(D and C)

= P(D | A) * P(A) + P(D | B) * P(B) + P(D | C) * P(C)

= 0.02 * 0.2 + 0.01 * 0.3 + 0.006 * 0.5

= 0.004 + 0.003 + 0.003

= 0.01

So, the probability of a phone being defective is 1%.

2. The probability that this defective phone is manufactured at factory A: P(A | D) = P(D and A) / P(D)

= (P(D | A) * P(A)) / P(D)

= (0.02 * 0.2) / 0.01

= 0.4

So, the probability that this defective phone is manufactured at factory A is 40%.

3. The probability that this defective phone is manufactured at factory B: P(B | D) = P(D and B) / P(D)

= (P(D | B) * P(B)) / P(D)

= (0.01 * 0.3) / 0.01

= 0.3

So, the probability that this defective phone is manufactured at factory B is 30%.

4. The probability that this defective phone is manufactured at factory C: P(C | D) = P(D and C) / P(D)

= (P(D | C) * P(C)) / P(D)

= (0.006 * 0.5) / 0.01

= 0.3

So, the probability that this defective phone is manufactured at factory C is 30%.


5. Explain the standard steps to perform machine learning.




**Answer:**

Machine learning is a complex process that involves several steps to ensure a successful outcome. Here are the standard steps in performing machine learning:

1. Define the problem: The first step is to define the problem you want to solve using machine learning. This involves understanding the problem domain, identifying the target audience, and defining the objectives of the model.

2. Collect and clean data: Once the problem is defined, you need to collect and clean the data that will be used to train the model. This involves acquiring data from various sources, preprocessing the data, and removing any missing or irrelevant information.

3. Exploratory Data Analysis (EDA): After cleaning the data, you need to perform Exploratory Data Analysis (EDA) to understand the structure of the data and identify any patterns, trends or relationships. This step is crucial for selecting the appropriate algorithm for your problem.

4. Feature Engineering: Feature engineering involves creating new features from existing ones to improve the performance of the machine learning model. This step involves feature selection, feature scaling, and feature normalization.

5. Model selection: Once the features are engineered, you need to select the appropriate machine learning algorithm for the problem. The choice of algorithm depends on the type of problem, the size and nature of the data, and the computational resources available.

6. Model Training: After selecting the algorithm, the next step is to train the model on the cleaned and processed data. The model is trained using an optimization algorithm that minimizes the error between the predicted values and the actual values.

7. Model evaluation: Once the model is trained, you need to evaluate its performance using appropriate evaluation metrics. This step is crucial to identify any overfitting or underfitting of the model and to determine if further improvement is required.

8. Hyperparameter tuning: If the model is not performing well, you need to tune its hyperparameters to improve its performance. Hyperparameters are the parameters that control the behavior of the model and are set before training the model.

9. Deployment: Once the model is trained and its performance is satisfactory, you can deploy it to make predictions on new data.

7. Write the objective function that can be used to determine the regression model parameters. How is this objective function will be used to find model parameters?




**Answer:**

The objective function that is commonly used for determining the regression model parameters is the mean squared error (MSE) function. The MSE function is defined as the average of the squared differences between the predicted values and the actual values in the training data. Mathematically, it can be expressed as:

$MSE$ = $\sum_{i=1}^{n} $ ($y_i - ŷ_i)^2$

Where $y_i$: true value, $ŷ_i$: output value and n: sample (dataset) size

To find the model parameters, we need to minimize the MSE function with respect to the model parameters. This is typically achieved using optimization algorithms such as gradient descent or numerical optimization methods.
The optimization process starts with initializing the model parameters with some values and iteratively adjusting the parameters until the MSE function is minimized. In each iteration, the gradient of the MSE function with respect to the parameters is computed and used to update the parameters in the direction that minimizes the MSE. This process continues until the MSE reaches a minimum or a stopping criterion is met. The model parameters obtained after the optimization process are the final parameters that minimize the MSE and are used for making predictions on new data.
